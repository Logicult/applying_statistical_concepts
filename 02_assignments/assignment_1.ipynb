### Question 1: Simple Linear Regression 

Let's set up our workspace and use the `Boston` dataset in the `ISLP` library. Print `Boston` to learn more about the dataset.

```python
# Import standard libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import subplots
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import specific objects
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from statsmodels.stats.outliers_influence \
     import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

```python
# Load the "Boston" dataset using the "load_data" function from the ISLP package
Boston = load_data('Boston')
```

Before we fit and review model outputs, we should visualize our data. Review the code and plot, shown below. Answer the following questions:

_(i)_ What are the `medv` and `dis` variables being plotted? (Hint: review this [link](https://islp.readthedocs.io/en/latest/datasets/Boston.html))

_(ii)_ What concept defines the plotted line?

(i) "medv' is the median value of owner-occupied homes in 1000's (USD).
    'dis' variable is the weighted mean distance from homes to Boston employment centers.

(ii) The dependant (response) variable is 'medv' and the independant (predictor) variable is 'dis' in the plotted graph. The plotted line shows a linear regression between the 2 variables. The plotted line shows the median value of owner-occupied homes for different distances to Boston employment centers.

```python
# Extract the variables
medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)

# Plot data
plt.scatter(dis, medv, label='Data')
plt.xlabel('dis')
plt.ylabel('medv')

# Fit a linear regression model
lm = LinearRegression()
lm.fit(dis, medv)

# Plot the regression line
plt.plot(dis, lm.predict(dis), color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('dis')
plt.ylabel('medv')
plt.legend()

# Show the plot
plt.show()
```

Consider the variables plotted above. In the context of the `Boston` dataset:

_(iii)_ What is the (implied) null hypothesis? What is the (implied) alternative hypothesis?

_(iv)_ Now, lets fit a simple regression model, using the general syntax `sm.OLS()` and `model.fit()`. As above, use `medv` as the response variable Y, and `dis` as the predictor variable X. (Hint: use the syntax `sm.add_constant` with the appropriate argument).

(iii) The (implied) null hypothesis would be that there is no significant statistical relationship between the 2 variables 'medv' (median value of owner-occupied home in 1000's) and 'dis' (weighted mean distance of home from Boston employment centers). The (implied) alternative hypothesis will be opposite as in there seems to be a significant statistical relationship between the two varibales.

```python
# Add your code here
# Answer to (iv)

# Extract the variables
medv = Boston['medv']
dis = Boston['dis']

# Add constant
X = sm.add_constant(dis)

# Fit model
model = sm.OLS(medv, X).fit()

# Summary
print(model.summary())
```

(vi) Standard errors for B0 and B1 are 0.817 and 0.188 respectively.
(vii) Confidence intervals for B0 and B1 are (16.784, 19.996) and (0.722, 1.462) respectively at 95% as shown in table.


Now, let's interpret the model output.

### Question 2: Multiple Linear Regression 

We'll continue to use the `Boston` dataset for this question.

_(i)_ Fit a multiple linear regression, with two predictor variables: $X_1$ is `dis`, and $X_2$ is `rm`. As before, keep `medv` as the response variable Y. (Hint: use the syntax `sm.add_constant` with the appropriate arguments).

```python
# Your answer here

medv = Boston['medv']
dis = Boston['dis']
rm = Boston['rm']

# Array
X = np.column_stack((dis, rm))

X = sm.add_constant(X)
model = sm.OLS(medv, X).fit()
print(model.summary())
```

_(ii)_ In the context of the `Boston` dataset, state the null and alternative hypotheses.

_(iii)_ Review the model output, using `summary()`. Does it appear that both `dis` and `rm` are predictive of `medv`? How did you determine this?

_(iv)_ We can use the inbuilt `sm.graphics.plot_regress_exog` function to generate helpful diagnostic plots (Hint: provide `plot_regress_exog` with the multiple regression model). Review the first generated plot, 'Residuals vs. Fitted'. Which observations are outliers? What impact might outliers have on our model?

(ii) The null hypothesis will be that there is no statistically significant relationship between median value of Owner-occupied homes  (medv) and both response variables weighted mean of distance from Boston employment centers (dis) and avergae number of rooms per dwelling (rm).
    The alternative hypothesis will be opposite of null hypothesis which is that at least one of the response variables has a statistically significant relationship with the predictive variable (medv)


(iii) As both p-values of dis (x1) and rm (x2) are less than 0.05, we ca safely say that both predictor variables have a statistically significant relationship with medv. They are both predictive of the response variable.

```python
# Add your code here
# Answer to (iv)

# Diagnostic plots for x1
fig = sm.graphics.plot_regress_exog(model, 'x1')
plt.show()

# Diagnostic plots for x2
fig = sm.graphics.plot_regress_exog(model, 'x2')
plt.show()
```

_(v)_ Fit a second model, this time including an interaction between the two predictor variables. Is there an interaction? (Hint: add a variable `x1 * x2` where `x1` and `x2` are the predictor variables). State an interpretation of the interaction, in the context of the `Boston` dataset, in one or two sentences.

Interaction is dsplayed in row with variable X3. As the p-value is less than 0.05 we can safely say that there is a statistically significant interaction between dis and rm predictor variables. This means that the correlation of either of the predictor variables with medv changes with the change in other predictor variable.

```python
# Add your code here
# Interaction
interaction = dis * rm

# dis + rm + interaction
X_interaction = np.column_stack((dis, rm, interaction))

X_interaction = sm.add_constant(X_interaction)
model_interaction = sm.OLS(medv, X_interaction).fit()
print(model_interaction.summary())
```
